## Introducción

**¿Sabías que los juegos para niños, pueden verse como vectores?**\
Cuando se tienen 2 o más columpios, estos pueden verse como vectores paralelos, ya que estos se mueven en una misma dirección, más no con la misma magitud, ya que el movimiento de ellos depende de la fuerza que se les aplique y el tiempo en el que se aplica dicha fuerza. Dentro de este notebook, aprenderemos qué es un vector, así algunas operaciones con estos.

El término vector proviene del latín vector, vectoris, cuyo significado es 'el que conduce', o 'el que transporta'.
Se le denomina vector de dimensión $n$ a una tupla de números reales (componentes del vector). El conjunto de todos los vectores de dimensión $n$, se representa como $R^n$. Es decir, un vector $\vec{v}$ perteneciente a un espacio vectorial $R^n$ se representa de la siguiente manera:\
$\vec{v}=(a_{1},a_{2},a_{3},...,a_{n})$, donde $\vec{v} \in R^n$

   + **¿Qué es un vector?**\
Dentro del área de la física, se le conoce como vector a un segmento de recta en el espacio que parte de un punto hacia otro, es decir, que tiene dirección y sentido. Los vectores en física tienen por función expresar las llamadas magnitudes vectoriales. En computación, un vector representa una lista de números que pueden tomar muchos valores.
En matemáticas se define vector como un elemento de un espacio vectorial. Esta noción es más abstracta y para muchos espacios vectoriales no es posible representar sus vectores mediante el módulo y la dirección, en particular los espacios de dimensión infinita sin producto escalar no son representables de ese modo. Los vectores en un espacio euclídeo se pueden representar geométricamente como segmentos de recta $R$, en el plano $R^{2}$ (bidimensional), o en el espacio $R^{3}$ (tridimensional).

Mientras en el área de la computación, los vectores son una forma de almacenar datos que permiten contener una serie de valores del mismo tipo, cada uno de los valores contenidos tiene una posición asociada que se usará para accederlos. Está posición o índice será siempre un número entero positivo.

# Suma y resta de vectores

Para sumar dos vectores $\vec{a}$  y $\vec{b}$ se suma $\vec{a}$ con el vector $\vec{b}$, es decir, se suman los componentes de cada vector:

\begin{equation}
\vec{a}+\vec{b} = (a_{x}+b_{x},a_{y}+b_{y},a_{z}+b_{z})
\end{equation}


### Ejemplo

Sean $\vec{a}=(3,2,-4)$ y $\vec{b}=(-3,2,7)$, calcula el vector $\vec{a}+\vec{b}$
\begin{equation}
\vec{a}+\vec{b} = (3+(-3),2+2,-4+7)=(0,4,3)
\end{equation}

Para restar dos vectores $\vec a$ y $\vec b$ se suma $\vec a$ con el opuesto del vector $\vec b$, es decir:}
\begin{equation}
\vec a- \vec b = (\vec a_{x}- \vec b_{x}, \vec a_{y}- \vec b_{y},\vec a_{z}- \vec b_{z})
\end{equation}
Por ejemplo:\
Sea $\vec a=(5,2,4)$ y $\vec b=(-3,5,9)$, calcular el vector $\vec a- \vec b$
\begin{equation}
\vec a- \vec b=(5-(-3),2-5,4-9)=(8,-3,-5)
\end{equation}

Sumando y Restando Vectores, al vector $\vec x$ se le desea sumar otro vector $\vec y$

```
#Declaramos nuestro vector y
y = np.array([2, -1,])
print('Nuestro vector x es:\n', x)
print('Nuestro vector y es:\n', y, '\n')
#Utilizando los operadores aritmeticos + y -, realizamos las operaciones necesarias
print(f'La suma de nuestros vectores X y Y es: {x+y}')
print(f'La resta de nuestros vectores X y Y es: {x-y}')
```

# Multiplicación de un Vector por un escalar

La multiplicación por un escalar proviene de la definición ya que este valor por el cuál vamos a multiplicar un vector en realidad solo va a agrandar ó a dismininuir la magnitud del vector (lo va a escalar).

## Ejemplo

Al multiplicar un vector $\vec{x}$ por un escalar $\lambda$, obtenemos como un nuevo vector $\vec{y}=(\lambda)(\vec{x})$ que cumple con las siguientes características:\
Si $\lambda$ es:\
\
~ Positivo $\vec{x}$ y $\vec{y}$ tendrán el mismo sentido.\
~ Negativo $\vec{x}$ y $\vec{y}$ tendrán diferente sentido.\
\
Por ejemplo:\
Sea $\vec{x} =(2,4,1,2)$ y $\lambda=2$\
\begin{equation}
(\vec{x})(\lambda) = 2(2,4,1,2) = (4,8,2,4)
\end{equation}

No importa cuantas componentes tenga nuestro vector, el escalar $\lambda$ multiplica cada una de las componentes del vector. Un ejemplo en 2 dimensiones es el siguiente.

Para multiplicar un vector $\vec{x}$ por un escalar $\lambda$, únicamente definimos dicho escalar y realizamos el producto con nuestro arreglo.

```
#Asignamos una variable k representando nuestro escalar
k = 2
#Declaramos nuestro vector
vector = np.array([1, 3])
#Realizamos la multiplicación
producto = vector*k
print('Vector: ', vector, '\n')
print('Escalar: ', k, '\n')
print(f'El producto es: {producto}')
```

Cuando el escalar es positivo

```
par = [{'title':'Producto (+Escalar)(Vector)','ylim':(-3,9), 'xlim':(-0.5,3.5)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 1]), y=[0,3], par={'label':'Vector', 'ls':'dashdot','lw':1, 'color':'blue'})
graf.plot(x=np.array([0, 2]), y= [0,6], par={'label':'Producto', 'ls':':','lw':2, 'color':'red'})
graf.scatter(x=np.array([0,1]), y=[0,3], par={ 'ls':'dashdot','lw':4, 'color':'blue','marker':'v'})
graf.scatter(x=np.array([0, 2]), y=[0,6], par={ 'ls':':','lw':6, 'color':'red','marker':'v'})
graf.legend()
graf.show()
```

Cuando el escalar es negativo, se tienen direcciones distintas

```
#Asignamos un valor para nuestro escalar
k = -2

#Realizamos la operación indicada
producto = vector*k
print('Escalar: ',k)
print(f'El producto es: {producto}')

par = [{'title':'Producto (+Escalar)(Vector)','ylim':(-3,9), 'xlim':(-2.0,2.0)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 1]), y=[0,3], par={'label':'Vector', 'ls':'dashdot','lw':1, 'color':'blue'})
graf.plot(x=np.array([-1, 0]), y=np.array([-2, 0]), par={'label':'Producto', 'ls':':','lw':2, 'color':'red'})
graf.scatter(x=np.array([0,1]), y=[0,3], par={ 'ls':'dashdot','lw':4, 'color':'blue','marker':'v'})
graf.scatter(x=np.array([0, 2]), y=[0,6], par={ 'ls':':','lw':6, 'color':'red','marker':'v'})
graf.legend()
graf.show()
```

# Vector Unitario
Un vector unitario hace referencia al vector cuyo módulo (distancia con respecto al origen) es igual a 1. Cabe recordar que el módulo es la cifra coincidente con la longitud cuando el vector se representa en un gráfico. El módulo, de este modo, es una norma de la matemática que se aplica al vector que aparece en un espacio euclídeo.
Para hallar un vector unitario a partir de cualquier vector, hay que dividir este último por su módulo.
### Ejercicio:
Si $\vec{v}$ es un vector de componentes (3,4), encontrar un vector unitario de su misma dirección y sentido, dada la fórmula:
\begin{equation}
\vec{u} = \frac{1}{|\vec{v}|} \vec{v}
\end{equation}
- Calcular magnitud de $\vec{v}$
\begin{equation}
|\vec{u}| = \sqrt{(3)^2 + (4)^2} = \sqrt{25} = 5
\end{equation}
Sustituyendo:
\begin{equation}
\vec{u} = \frac{1}{5}(3,4) = (\frac{3}{5},\frac{4}{5})
\end{equation}
Visto desde código:

```
# Creamos una lista vacía para nuestro vector unitario y otra con valores para nuestro vector

vector_unitario = []
vector = [3,4]
#Calculando magnitud
magnitud = math.sqrt(vector[0]**2 + vector[1]**2)
#En el siguiente ciclo, almacenamos los valores de nuestro vector
for i in range(len(vector)):
  vector_unitario.append(vector[i]*(1/magnitud))
print('Vector original', vector)
print('Nuestro vector unitario es: ',vector_unitario)
```

```
par = [{'title':'Vector Unitario','ylim':(0,4.5), 'xlim':(0,3.5)}]
graf = vis.planoCartesiano(par=par)
graf.plot(x=np.array([0, 3]), y=[0,4], par={'label':'Vector', 'ls':'dashdot','lw':1, 'color':'blue'})
graf.plot(x=np.array([0, 0.6]), y= [0,0.8], par={'label':'Vector Unitario', 'ls':':','lw':2, 'color':'red'})
graf.scatter(x=np.array([0, 3]), y=[0,4], par={ 'ls':'dashdot','lw':4, 'color':'blue','marker':'v'})
graf.scatter(x=np.array([0, 0.6]), y=[0,0.8], par={ 'ls':':','lw':6, 'color':'red','marker':'v'})
graf.legend()
graf.show()
```

$$|d| = \sqrt{1825}$$
