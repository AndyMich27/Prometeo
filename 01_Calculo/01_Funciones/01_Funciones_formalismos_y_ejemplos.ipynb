De manera más formal:

Sean $A$ y $B$ dos conjuntos. Una función $f$ de $A$ en $B$ es una relación $f \subseteq A$ x $B$ tal que:

(i) para toda $x \in A$, existe $y \in B$ tal que $(x,y) \in f$, es decir,
$\forall x(x \in A \Rightarrow (\exists y (y \in B \wedge (x,y) \in f)))$; y

(ii) Cada elemento de $A$ tiene asociado sólo un elemento de $B$, es decir, $\forall x \in A$ $\forall y_1,y_2 \in B [((x,y_1) \in f \wedge (x,y_2) \in f) \Rightarrow y_1 = y_2 ]$



Juntando las condiciones (i) y (ii), podemos decir que para toda $x \in A$
existe una única $y \in B$ tal que $(x,y) \in f$ . Entonces, si $f$ es una función de $A$ en $B$, para cada $x \in A$ denotamos con $f(x)$ al único elemento de $B$ que le corresponde. Es decir, si $(x,y) \in f$, entonces escribimos $f(x) = y$. A $f(x)$ le llamamos la imagen de $x$ bajo $f$. En ocasiones, decimos que el elemento $x \in A$ es mandado bajo la función $f$ a $f(x)$ y lo denotamos por $x \mapsto f(x)$.
Como una función $f : A \rightarrow B$ es una relación binaria entre $A$ y $B$, la definición de la imagen de f es:

$im(f) = \{b \in B : \exists a \in A \wedge $(a,b)$ \in f\}$,
pero entonces, como $f$ es función,

$im(f)= \{y \in B : \exists x \in A f(x) = y \} = \{f(x) : x \in A\}$.
Por lo tanto, $im(f) \subseteq  B$.

## Propiedades de las funciones

Sean $A, B$, $C$ y $D$ conjuntos cualesquiera. Sean
$f: A \rightarrow B$, $g:C \rightarrow D$. Definimos la siguientes operaciones entre $f$ y $g$:

1. La suma de $f$ y $g$ mediante la ecuación: $(f + g)(x) = f(x) +g(x)$, donde
 $f+g:A\cap C \rightarrow B\cap D$ 
2. La resta de $f$ y $g$ mediante la ecuación: $(f - g)(x) = f(x) -g(x)$, donde
 $f-g:A\cap C \rightarrow B\cap D$ 
3. El producto de $f$ y $g$ mediante la ecuación: $(f \cdot g)(x) = f(x)\cdot g(x)$. Particularmente, si $f(x) = k$, definimos una nueva función $k\cdot g$ como $(k\cdot g)(x) = k\cdot g(x)$. Por lo tanto, $f \cdot g: A \cap C \rightarrow B\cap D$.
4. El cociente de $f$ y $g$ mediante la ecuación: $(\frac{f}{g})(x) = \frac{f(x)}{g(x)}$, siempre que $g(x) \neq 0$ para toda $x \in A$. Por lo tanto, $\frac{f}{g}: A \cap C \cap \{x|g(x)\neq 0\} \rightarrow B\cap D$.


**Teorema** Sean $A, B$, $C$,$D$, $E$ y $F$ conjuntos cualesquiera. Sean
$f:A\rightarrow B$, $g:C \rightarrow D$ y $h:E \rightarrow E$. Se cumple lo siguiente:

1. *Asociatividad*: $(f+g)+h=f+(g+h)$  y  $(f\cdot g)\cdot h=f\cdot (g\cdot h)$
2. *Conmutatividad*: $f+g=g+f$  y  $f\cdot g=g\cdot h$


## Tipos de Funciones

Una función $f : A \rightarrow B$ es **inyectiva** o **uno a uno** si y
sólo si para cualesquiera $x_1, x_2 \in A$ con $x_1 \neq x_2$ se tiene que $f(x_1) \neq f(x_2)$, es decir:
$\forall x_1,x_2 \in A(x_1 \neq x_2 \Rightarrow f(x_1) \ne f(x_2))$.
Equivalentemente, usando la contrapuesta, $f$ es inyectiva si y sólo si:

$\forall x_1,x_2 \in A(f(x_1) = f(x_2) \Rightarrow x_1 = x_2)$.

En otras palabras, una función es inyectiva si elementos distintos del dominio van a dar a elementos distintos del codominio al aplicar la función, o bien si la única manera de que $x_1$ y $x_2$ en el dominio vayan a dar al mismo elemento del codominio sea que desde un principio $x_1 = x_2$.

Una función $f : A \rightarrow B$ es **sobreyectiva** o **suprayectiva** si y sólo
si para todo $y \in B$ existe $x \in A$ tal que $f(x) = y$ , es decir,
$\forall y (y \in B \Rightarrow \exists  x \in A \wedge f(x) = y$.
De acuerdo a lo anterior las funciones suprayectivas son aquellas en que todo elemento del codominio proviene de alguno del dominio.

Una función $f : A \rightarrow B$ es **biyectiva** si y sólo si $f$ es
inyectiva y es sobreyectiva.

### Ejemplos

1. Sean $A = \{-1, 1, 2,3\}, B = \{-8, 4, 5, 8\}$ y $f:A\rightarrow B$
definida como $f(x) = -x^2 +2x +7$ .
Podemos ver que $f$ no es inyectiva, pues $f(-1) = 4$ y $f(3) = 4$, pero $-1 \neq 3$.
Además, $f$ no es sobreyectiva, pues $5 \in B$ y no hay $a \in A$ tal que $f(a) = 5$. Sin embargo, si escogemos a los conjuntos $C$ y $D$ de tal forma que $C=A-\{3\}$ y $D=B-\{5\}$, la función $f:C\rightarrow D$ definida como $f(x) = -x^2 +2x +7$ sí es sobreyectiva e inyectiva, es decir, $f(x)$ sería biyectiva.

**NOTA:** Siempre es posible restringir el dominio de una función para crear una función inyectiva y el codominio para crear una función suprayectiva.

2. Sea $f : \mathbb{R^+} \rightarrow (-8,\infty)$ dada por $f(x) = 9x -8$.
Veamos que $f$ es biyectiva.
Sean $x, y  \in \mathbb{R^+} $ tales que $f(x) = f(y)$. Entonces por la ley de la cancelacióón de la suma: $9x -8 = 9y -8 \Longleftrightarrow 9x = 9y \Longleftrightarrow x = y$. Por lo tanto, $f$ es inyectiva. 
Ahora bien, sea $y \in (-8,\infty)$. Necesitamos eoncontrar $x \in \mathbb{R^+}$ tal que $f(x)=y$. Dado que $f(x)=9x -8$, necesitarı́amos que $9x -8= y$, entonces $x =\frac{y + 8}{9}$. Ası́, sea $x =\frac{y + 8}{9}$, entonces $f(x) = f (\frac{y + 8}{9}) = 9(\frac{y + 8}{9}) -8 = (y+8)-8= y$. Por lo tanto, para cualquier $y \in (-8,\infty)$, hay  $x \in \mathbb{R^+}$  tal que $f (x) = y$ y $f$ es sobreyectiva.
Por lo tanto, $f$ es biyectiva.

# Composición de Funciones

Sean $A, B$ y $C$ conjuntos cualesquiera. Sean $B' \subseteq B$,
$f:A\rightarrow B'$ y $g:B \rightarrow C$. Definimos la **composición** de $f$ y $g$, denotada por $g \circ f$ , como la función $g \circ f : A \rightarrow C$ tal que para toda $x \in A, (g \circ f )(x) =
g(f(x))$.

## Ejemplo
Sean $f : \mathbb{R} \rightarrow  \mathbb{R}$ y $g : \mathbb{R} \rightarrow \mathbb{R} $ definidas por $f (x) = 5x^2 + 4$
y $g(x) = 8x -3$.
Entonces la composición $g \circ f: \mathbb{R} \rightarrow \mathbb{R}$ está dada por $(g \circ f )(x) =
g(f (x)) = g(5x^2 + 4) = 8(5x^2 + 4) -3 = 40x^2 + 32 -3 = 40x^2 + 29 $.

La composición $f \circ g: \mathbb{R} \rightarrow \mathbb{R}$ está dada por $(f \circ g)(x) = f (g(x)) =
f (8x -3) = 5(8x -3)^2 + 4 = 5(64x^2 - 48x +9)+4 = 320x^2 -240x + 45 + 4 =  320x^2 -240x + 49$.

Es decir, la composición de funciones no es conmutativa, pues
$(g \circ f)(2) = 40x^2 + 29  = 189 \neq 849 =  320(2)^2 -240(2) + 49 = (f \circ g)(2)$,
por lo que $g \circ f \neq f \circ g$.

```python
def f5(x):
    return 5 * x**2 + 4

def g5(x):
    return 8*x - 3 

x5 = np.arange(-10, 10, 0.1) 

f_g = f5( g5(x5) ) 
g_f = g5( f5(x5) )

par = [{'title':'Composición de funciones', 
        'xlabel':'$X_5$'},       
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x = x5, y = f_g, par={'label':'$f(g(x))$'})
graf.plot(x = x5, y = g_f, par={'label':'$g(f(x))$'})
graf.legend()
graf.show()
```

**Teorema.** Sean $A, B, C$ y $D$ conjuntos cualesquiera, y sean $B' \subseteq B$
y $C \subseteq C$. Sean $f : A \rightarrow B'$ , $g : B \rightarrow C'$ y $h : C \rightarrow D$ funciones. Entonces
$h \circ (g \circ f ) = (h \circ g) \circ f$ , es decir, la composición de funciones es asociativa.

**Eejmplo***

Sean Sean $f : \mathbb{R} \rightarrow \mathbb{R}$ , $g : \mathbb{R}\rightarrow \mathbb{R}$ y $h : \mathbb{R} \rightarrow \mathbb{R}$ funciones dadas por las siguientes reglas de correspondencia: $f(x) = e^x$, $g(x) = 3x^2$ y $h(x) = sin(x)$.

Por un lado, tenemos que: 

$h \circ (g \circ f )(x) = h \circ (g(f(x)) = h \circ g(e^x) = h( g(e^x)) = h( g(e^x)) = h(3e^{2x}) = sin(3e^{2x})$

Por otro lado:

$(h \circ g) \circ f = h(3x^2) \circ f = sin(3x^2) \circ f = sin(3(e^x)^2) = sin(3e^{2x})$

$\therefore h \circ (g \circ f ) = (h \circ g) \circ f$ 





```python
def h6(x):
    return np.sin(x)

def g6(x):
    return 3*x**2

def f6(x):
    return np.exp(x)

def h_g(x):
    return h6(g6(x))

def g_f(x):
    return g6(f6(x))

x6 = np.arange(-3, 3.1, 0.1) 
h_g_f_1 = h6( g_f(x6) )
h_g_f_2 = h_g( f6(x6))

par = [{'title':'$h \circ (g \circ f)$', 
        'xlabel':'$x$',
        'ylabel':'$y$'},
       {'title':'$(h \circ g ) \circ f$', 
        'xlabel':'$x$',
        'ylabel':'$y$'}, 
      ]

graf = vis.planoCartesiano(1, 2, par=par)
graf.plot(1, x = x6, y = h_g_f_1, par={'color':'C0'})
graf.plot(2, x = x6, y = h_g_f_2, par={'color':'C1'})
graf.show()
```

 
**Teorema**

$(i)$ La composición de funciones inyectivas es inyectiva.

$(ii)$ La composición de funciones sobres es sobre (siempre que el codominio de la primera sea igual al dominio de la segunda).

$(iii)$ La composición de funciones biyectivas es biyectiva.

**Ejemplo**

Sean $f:\mathbb{R^+}\rightarrow (-8,\infty), g:\mathbb{R^+}\rightarrow (\sqrt{3},\infty)$ dadas por $f(x) = 9x -8$ y $g(x)=\sqrt{7x+3}$.
En un ejercicio anterior vimos que $f$ es biyectiva. Veamos ahora que $g(x)$ es biyectiva:

Sean $x, y  \in \mathbb{R^+}$ tales que $g(x) = g(y)$. Entonces $\sqrt{7x+3} = \sqrt{7y+3} \Longleftrightarrow 7x+3 = 7y+3 \Longleftrightarrow 7x = 7y \Longleftrightarrow x = y$. Por lo tanto, $g$ es inyectiva. 
Ahora bien, sea $y \in (\sqrt{3},\infty)$. Necesitamos encontrar $x\in \mathbb{R^+}$ tal que $g(x)=y$. Dado que $g(x)=\sqrt{7x+3}$, necesitarı́amos que $\sqrt{7x+3}= y$, y esto pasa si y solo si $x= \frac{y^2-3}{7}$. Ası́, sea $x=\frac{y^2-3}{7}$, entonces $g(x) = g( \frac{y^2-3}{7}) = \sqrt{7(\frac{y^2-3}{7})+3} = \sqrt{(y^2-3) +3}=\sqrt{y^2} =y$ (pues $ y >0$). Por lo tanto, para cualquier $y\in (\sqrt{3},\infty$, hay  $x \in \mathbb{R^+}$  tal que $g(x) = y$ y entonces $g$ es sobreyectiva.
Por lo tanto, $g$ es biyectiva.
Por lo tanto $f$ y $g$ son biyectivas.

Por otro lado, verifiquemos que la composición de funciones biyectivas es biyectiva. Para ello, la ejemplificaremos con la composición $ f \circ g$.

$ f \circ g = f(g(x)) = f(\sqrt{7x+3}) = 9\sqrt{7x+3}-8$. Verifiquemos que esta composición de funciones biyectivas es biyectiva:

Sean $x, y  \in \mathbb{R^+} $ tales que $f(g(x)) = f(g(y))$. Entonces $9\sqrt{7x+3}-8 = 9\sqrt{7y+3}-8 \Longleftrightarrow 9\sqrt{7x+3} = 9\sqrt{7y+3} \Longleftrightarrow  \sqrt{7x+3} = \sqrt{7y+3} \Longleftrightarrow  7x+3 = 7y+3 \Longleftrightarrow  7x = 7y \Longleftrightarrow x = y$. Por lo tanto, $f(g(x))$ es inyectiva.
En cuanto a la sobreyectividad, sea $y \in (\sqrt{3},\infty) \subseteq \mathbb{R^+}$. Necesitamos encontrar $x \in \mathbb{R^+}$ tal que $f(g(x))=y$. Como $f(g(x))=9\sqrt{7x+3}-8$, se requiere que $9\sqrt{7x+3}-8= y$, y esto pasa si y solo si $x= \frac{(\frac{y+8}{9})^2 - 3}{7}$. Ası́, sea $x= \frac{(\frac{y+8}{9})^2 - 3}{7}$, entonces $f(g(x)) = f(g(\frac{(\frac{y+8}{9})^2 - 3}{7})) =  9\sqrt{7(\frac{(\frac{y+8}{9})^2 - 3}{7})+3}-8 = 9\sqrt{[(\frac{y+8}{9})^2 - 3]+3}-8 = 9\sqrt{(\frac{y+8}{9})^2}-8 = 9(\frac{y+8}{9})-8 = (y+8) - 8 = y$. 

De esta manera, para cualquier $y\in(\sqrt{3},\infty)$, hay  $x\in \mathbb{R^+}$  tal que $f(g(x)) = y$ y entonces $f(g(x))$ es sobreyectiva.
Por lo tanto, $f(g(x))$ es biyectiva, lo cual quiere decir que la composición de funciones biyectivas es biyectiva.








```python
def f7(x):
    return 9*x-8 

def g7(x):
    return np.sqrt(7*x+3)

x7 = np.arange(0, 40, 0.1) 

f2_g2 = f7( g7(x7) )

par = [{'xlabel':'$x$',
        'ylabel':'$y$'}, 
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x = x7, y = f2_g2, par={'label':'$f(g(x))$'})

N = 40
graf.plot(x = [-3,40], y = [N,N], par={'label':'Recta horizontal que corta en un solo punto',
                                           'ls':'--','lw':2, 'color':'green'})

graf.legend()
graf.show()
```

### Funciones inversas

Sea $f : A \rightarrow B$ una función. Un inverso izquierdo (derecho) de $f$ es una función $g: B \rightarrow A$ que $g \circ f = id_A (f \circ g = id_B)$.

El siguiente teorema da la condición necesaria y suficiente para que $f$
tenga inverso derecho (inverso izquierdo).

**Teorema** Sean $A$ y $B$ conjuntos cualesquiera y sea $f : A \rightarrow B$. Se cumple lo siguiente:

$(i)$ Si $A \neq \emptyset$, entonces $f$ es inyectiva si y sólo si $f$ tiene inverso izquierdo.

$(ii)$ $f$ es sobreyectiva si y sólo si $f$ tiene inverso derecho.

Es hacer notar que si $f$ es inyectiva, dependiendo de la elección de $a_0 \in A$, $f$ puede tener varios inversos derechos distintos, además de que puede no tener ningún inverso izquierdo. De manera análoga, si $f$ es sobreyectiva, dependiendo de la elección de cada $a_1 \in A$, $f$ puede tener varios inversos izquierdos, además de que puede no tener ningún inverso derecho.

Veamos un ejemplo de funciones $f$ y $g$ tales que $g \circ f = id_{\mathbb{Z}}$ y $f \circ g \neq id_{\mathbb{Z}}$.
**Ejemplo** Sean $f : \mathbb{Z} \rightarrow \mathbb{Z}$ y $g: \mathbb{Z} \rightarrow \mathbb{Z}$  dadas por $f(n) = 2n$ y
$g(n) = \lceil \frac{n}{2} \rceil$. 
Veamos que $g \circ f = id_{\mathbb{Z}}$ , pero que $g \circ f \neq id_{\mathbb{Z}}$.

 
Sea $z \in \mathbb{Z}$, entonces $(g \circ f )(z) = g(f (z)) = g(2z) = \lceil \frac{2z}{2} \rceil = \lceil z \rceil = z$. Además de que el dominio y codominio de $g \circ f$ y de $id_{\mathbb{Z}}$ es ${\mathbb{Z}}$. Por lo tanto, $g\circ f = id_{\mathbb{Z}}$ .
Por otro lado, 7 $\in \mathbb{Z}$ y evaluémoslo en $f \circ g$. Entonces $(f \circ g)(7) =f (g(7)) = f (\lceil \frac{7}{2} \rceil) = f (3) = 2(3) = 6 \neq 7$. Por lo tanto, $f \circ g \neq id_{\mathbb{Z}}$.


Afortunadamente, si una función f tiene inverso derecho e inverso izquierdo, entonces son el mismo, y en consecuencia sólo tiene un inverso derecho y un inverso izquierdo (que son el mismo).

**Teorema** Si $f : A \rightarrow B$ tiene inveso derecho $g_1$ y también tiene inverso izquierdo $g_2$ , entonces $g_1 = g 2$.

Sea $f : A \rightarrow B$. Se dice que $f$ es **invertible** si y solo
si existe $g : B \rightarrow A$ tal que $g$ es inverso derecho de $f, g \circ f = id_A$ , y $g$ es inverso izquierdo de $f$ , $f \circ g = id_B$.

***Lema*** Sea $f : A \rightarrow B$
$(i)$ Si $f^{-1}$ es función, entonces $f \circ f^{-1} = id_B$ y $f^{-1} \circ f = id_A$.

$(ii)$ $f$ es invertible si y sólo si $f^{-1}$ es función.

$(iii)$ Si $f$ es invertible, el inverso derecho e izquierdo de $f$ es $f^{-1}$.

***Corolario*** Una función $f : A \rightarrow B$ es invertible si y sólo si es biyectiva.

**Observaciones**

1. Las gráficas de $ f(x) $y $f^{-1}(x)$  son simétricas respecto de la bisectriz del primer y tercer cuadrante.
2. Para construir o calcular la función inversa de una función cualquiera, se deben seguir los siguientes pasos:

  2.1. Se escribe la función con $x$ e $y$.

  2.2. Se despeja la variable $x$ en función de la variable $y$.

  2.3. Se intercambian las variables.

**Ejemplos** De un ejemplo anterior sabemos que $g : \mathbb{R^+} \rightarrow (\sqrt{3},\infty)$ dada $g(x)=\sqrt{7x+3}$  es biyectiva; y entonces, por el corolario anterior, $g$ es invertible; por lo que al realizar el procedimiento de la segunda observación anterior, hacemos lo siguiente:

- Cambiamos $g(x)$ por $y$ $\Rightarrow y = \sqrt{7x+3}$

- Despejamos a la variable $x \Rightarrow x = \frac{y^2 - 3}{7}$

- Cambiamos $x$ por $g^{-1}(x)$ y obtendremos la función inversa $\Rightarrow g^{-1}(x) = \frac{x^2 - 3}{7}$, con $g^{-1}: (\sqrt{3},\infty)\rightarrow \mathbb{R^+}$



 




```python
x8 = np.arange(0, 20, 0.1)
y8 = np.arange(np.sqrt(3), 11.95 + np.sqrt(3), 0.1) 

g_8 = g7(x8)
g_8_inv = (y8**2 - 3) / 7

par = [{'xlabel':'$x$',
        'ylabel':'$y$'}, 
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x = x8, y = g_8, par={'label':'$g(x)$'})
graf.plot(x = y8, y = g_8_inv, par={'label':'$g^{-1}(x)$'})

graf.plot(x = [0,20], y = [0,20], par={'label':'Simetría de la función y su inversa',
                                       'ls':'--','lw':1, 'color':'gray', 'zorder':0})
graf.legend()
graf.show()
```

### Ejemplo 

Consideremos la función $j : \mathbb{R} \rightarrow \mathbb{R}$ dada por $j(x) = x^2 -4$. Podemos notar que sí es función, pues a cada elemento del dominio le corresponde un único elemento del contradominio. Sin embargo, la función no se inyectiva, pues existen $3, -3 \in \mathbb{R}$  tales que $ j(3) = (3)^2-4=5=(-3)^2-4 = j(-3)$ y $3\neq -3$; y tampoco es sobreyectiva, pues no hay $x \in \mathbb{R}$ tal que $j(x) = -7$ y $-7 \in \mathbb{R}$. Entonces, como $j$ no es biyectiva, no tiene función inversa. Si quisiéramos forzar la condición para encontrar la inversa, es fácil ver que su inversa sería de la forma $j^{-1}(x) = \pm \sqrt{x+4}$, pero $j^{-1}(x)$ sería una relación y no una función!, ya que $j^{-1}(5) = 3$ y $j^{-1}(5) = -3$. Podemos verificarlo con la gráfica de $j^{-1}(x)$:

```python
def j_inversa_pos(x):
    return np.sqrt(x+4)

def j_inversa_neg(x):
    return np.sqrt(x+4) * -1

x9 = np.arange(-5,5,0.1)
y9 = np.arange(-4, 20, 0.1)

j_9 = x9**2 - 4
j_inv_pos = np.sqrt(y9 + 4)
j_inv_neg = -1 * np.sqrt(y9 + 4)

par = [{'xlabel':'$x$',
        'ylabel':'$y$'}, 
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x = x9, y = j_9, par={'label':'$j(x)$'})
graf.plot(x = y9, y = j_inv_pos, par={'label':'$j^{-1}(x)$', 'color':'red'})
graf.plot(x = y9, y = j_inv_neg, par={'color':'red'})

graf.plot(x = [-4,20], y = [-4,20], par={'label':'Simetría de la función y su inversa',
                                         'ls':'--','lw':1, 'color':'gray', 'zorder':0})
N = 4
graf.plot(x = [N, N], y = [20,-5], par={'label':'Recta vertical',
                                       'ls':'--','lw':2, 'color':'green', 'zorder':0})                                      
graf.legend()
graf.show()
```

## Ejemplo de aplicación

El comportamiento periódico, es decir aquel que se repite una y otra vez, es común en la naturaleza y en los seres humanos. El parpadeo resulta en la repetitiva regla de día, pues nos la áásamos abriendo y cerrando los ojos. Ciertas poblaciones de animales aumentan y disminuyen en una forma periódica  que  se  puede  predecir:  una  población  grande  agota  la  provisión  de  alimento, lo cual hace que  la  población  disminuya;  esto,  a  su  vez,  resulta  en  una  provisión  más  abundante de alimento, lo cual hace posible que la población aumente; y el patrón se repite una y otra vez. Otros  ejemplos  comunes  de  comportamiento  periódico  comprenden  el  movimiento  que  es  causado  por  vibración  u  oscilación.  Una  masa  suspendida  de  un  resorte  que  ha  sido  comprimido  y  luego  se  deja  vibrar  verticalmente  es  un  movimiento  simple.  Este  movimiento "de vaivén" también se presenta en fenómenos diversos como por ejemplo ondas de sonido, ondas de luz, corriente eléctrica alternante y estrellas que pulsan, por citar  sólo  algunos. 

Las funciones trigonométricas son idealmente apropiadas para modelar el comportamiento periódico. Una mirada a las gráficas de las funciones seno y coseno, por ejemplo, nos dice que estas funciones por sí solas exhiben comportamiento periódico. La Figura 1 muestra la gráfica de $f(t)=sen(t)$. Si consideramos $t$ como tiempo, vemos que a medida que el tiempo transcurre, $f(t)=sen(t)$  aumenta  y  disminuye  una  y  otra  vez.
La  Figura  2  muestra  que  el  movimiento de una masa en vibración en

Observe que la masa regresa a su posición original una y otra vez. Un **ciclo** es una vibración completa de un cuerpo, de modo que la masa de la Figura 2 completa un ciclo de su movimiento entre O y P. Nuestras observaciones acerca de la forma en que las funciones seno y coseno modelan el comportamiento periódico se resumen a continuación:

**MOVIMIENTO ARMÓNICO SIMPLES**

Si la ecuación que describe el desplazamiento $f(t)$ de un cuerpo en el tiempo $t$ es:

$f(t)=asen(\omega t)$ o $f(t)=acos(\omega t)$

entonces el cuerpo está en movimiento armónico simple. En este caso:

*$amplitud = |a|$* (**Desplazamiento máximo del cuerpo**)

*$periodo = \frac{2\pi}{\omega}$* (**Tiempo requerido para completar un ciclo**)

*$frecuencia = \frac{\omega}{2\pi}$* (**Número de ciclos por unidad de tiempo**)


**Ejemplo** Un  músico  que  toca  una  tuba  hace  sonar  la  nota  *Mi*  y  sostiene  el  sonido  durante  algún  tiempo. Para una nota *Mi* pura, la variación en presión a partir de la presión normal del aire está dada por: $V(t)=0.2sen(80\pi t)$, donde $V$ se mide en libras por pulgada cuadrada y $t$ se mide en segundos.

1. Encuentre la amplitud, período y frecuencia del movimiento de $V$.
2. Si el músico aumenta la intensidad de la nota, ¿cómo cambia la ecuación de V? 
3. Si el músico está tocando una nota incorrectamente y es un poco desafi nada, ¿cómo cambia la ecuación de V?.
4. Trace una gráfica de $V$.


**Solución** 

1. De la fórmula para amplitud, período y frecuencia obtenemos:

 > $amplitud=|a|=|0.2|=0.2$ $pulg.$

 > $periodo = \frac{2\pi}{\omega}=\frac{2\pi}{80\pi} = \frac{1}{40}s$

 > $frecuencia = \frac{\omega}{2\pi}= \frac{80\pi}{2\pi} = 40$ ciclos por segundo $(Hz)$.

2. Si el músico aumenta la intensidad, la amplitud aumenta. Por lo tanto, el número $0.2$ es sustituido por un número más grande.

3. Si la nota es desafinada, entonces la frecuencia disminuye. En consecuencia, el coeficiente de $t$ es menor a $80\pi$.

4. La gráfica de $V$ en el tiempo $t$ la mostramos en seguida:



```python
def V(t):
    return 0.2*np.sin(80*np.pi*t)

t = np.arange(0,0.5,0.001)
v = V(t)

par = [{'title':'Gráfica de la función  0.2 sen $(80\pi t)$',
        'xlabel':'$t$ [s]',
        'ylabel':'$V(t)$ [lb / in$^2$]'}, 
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x = t, y = v)                                     
graf.show()
```

A continuación, el usuario es capaz de indicar la función que quisiera graficas:

```python
import re

replacements = {
    'sin' : 'np.sin',
    'cos' : 'np.cos',
    'tan' : 'np.tan',
    'floor': 'np.floor',
    'ceil': 'np.ceil',
    'exp': 'np.exp',
    'sqrt': 'np.sqrt',
    'pi': 'math.pi',
    'e': 'math.e',
    '^': '**',
}

allowed_words = [
    'x',
    'sin',
    'cos',
    'tan',
    'ceil',
    'floor',
    'sqrt',
    'exp',
    'abs',
    'pi',
    'e'
]

def string2func(string):
    ''' evalúa una cadena y devuelve una función de x '''
    # encuentra todas las palabras y verifique si todas están permitidas::
    for word in re.findall('[a-zA-Z_]+', string):
        if word not in allowed_words:
            raise ValueError(
                '"{}" está prohibido usar en expresiones matemáticas'.format(word)
            )

    for old, new in replacements.items():
        string = string.replace(old, new)

    def func(x):
        return eval(string)

    return func
```

```python
#func_str = str(input('Introduzca la función que desea graficar:\n f(x) = '))
func_str = 'sin(e*x^2)'
func = string2func(func_str)
a = float(input('Introduzca el límite inferior: '))
b = float(input('Introduzca el límite superior: '))
x = np.linspace(a, b, 1000)

#print(str(func))

par = [{'title':'$'+func_str+'$','xlabel':'x','ylabel':'f(x)'},]
graf = vis.planoCartesiano(par=par)
graf.plot(x = x, y = func(x))
graf.show()
```
