## Introducción

Las derivadas son funciones matemáticas que, a partir del siglo XVII, gracias a los estudios de Isaac Newton y Leibniz, dieron solución al cálculo infinitesimal, que se había empezado a estudiar en la Grecia clásica, más o menos en siglo III a. C. 
Su importancia radica en la solución de millones de problemas matemáticos observados en la vida cotidiana. Incluso soluciona problemas y comportamientos de la naturaleza pues con ayuda de las derivadas es posible caracterizar el crecimiento de una población por ejemplo.
Así, las derivadas son esenciales para estudios tan importantes como el de la relatividad, la mecánica cuántica, la ingeniería, ecuaciones diferenciales, teoría de las probabilidades, sistemas dinámicos, teoría de las funciones, computación, etc.

Las derivadas aportan información concreta, directa y científica a los expertos y, con esos resultados, interpretan y son capaces de ofrecer información acerca de nuestra propia existencia y también utilizarlas para aplicarlas en cosas tan habituales como el vuelo de un avión, el movimiento de un coche, la construcción de un edificio, de un contenedor o de muchos otros elementos.


## Teoría
 ### Definición de derivada
 En una definición matemática formal pordemos definir la derivada de la siguiente manera:

 Existen muchas funciones para las cuales es posible encontrar el valor de la pendiente de una recta tangente a la función mediante el uso de una fórmula como la siguiente:
 $$ f´(x)=\lim_{h \to 0} \frac{f(x + h) - f(x)}{h} $$
 La función anterior se denomina la derivada de la función *f* siempre que el límite exista.

 La derivada de una función esta representada gráficamente como una línea recta superpuesta sobre cualquier función y como vimos en la definición, es el valor de esta pendiente respecto al eje. Por lo tanto intuitivamente podriamos decir que la derivada mide el ritmo de cambio de una variable respecto de otra.


 ### Reglas de derivación
 

*   Regla de potencias

 Para cualquier numero real *n* si $f(x)= x^n$

 
 *   Regla de la función constante

 Si $f(x)=c$ es una función constante, entonces
$$ f´(x)=o $$

 *   Regla de la multiplicación por constante

 Si *c* es cualquier constante y *f* es diferenciable en *x*, entonces *g=cf* es diferenciable en *x* y
 $$ g´(x)=c f(x) $$

 *   Regla de suma y diferencia

 Si *f* y *g* son diferenciables en *x*, entonces *f+g* y *f-g* son diferenciables en *x*,
 $$ [f(x)+ g(x)]´=f´(x) + g´(x) $$

 $$ [f(x)- g(x)]´=f´(x) - g´(x) $$

 *   Regla del producto

 Si *f* y *g* son funciones diferenciables en *x*, entonces $f*g$ es diferenciable en *x* 
$$ [f(x)*g(x)]´= f(x)g´(x) + g(x)f´(x) $$

 *   Regla del cociente

 Si *f* y *g* son funciones diferenciables en *x* y $g(x)\neq 0$ , entonces $f/g$ es diferenciable en *x* 
$$  \left[\frac{f(x)}{g(x)} \right] = \frac{f(x)g´(x)- f(x)g´(x) }{g(x)^2}$$

 *   Regla  de la cadena 

 Si la función *f* es diferenciable en *u=g(x)* y la función *g* es diferenciable en *x*, entonces la composición $y=(f\circ g)(x)= f(g(x)) es diferenciable en *x*
 $$ f(g(x))´= f´(g(x)) \cdot g´(x)  $$

*    Regla de L’Hôpital

 Esta regla es utilizada en caso de indeterminaciones donde *f y g* son dos funciones continuas definidas en el intervalo $[a,b]$, derivables en $(a,b)$ y sea *c* perteneciente a $(a,b)$ tal que $f(c)=g(c)=0$ y $g'(x)≠0 $ si $x≠c.$ Si existe el límite *L* de $f'/g'$ en *c*, entonces existe el límite de $f/g$ (en c) y es igual a L. Por lo tanto:
  $$ \lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f´(x)}{g´(x)} = L $$



 ### Derivadas de funciones trigonométricas
Las derivadas de las funciones trogonométricas son:
1.   $$sen´(x) = cos(x)$$
2.   $$tan´(x) = sec^2(x)$$
3.   $$sec´(x) = sec(x)tan(x)$$
4.   $$cos´(x) = -sen(x)$$
5.   $$cot´(x) = -csc^2(x)$$
6.   $$csc´(x) = -csc(x)cot(x)$$

### Teorema de Rolle
 Michel Rolle fue un matemático francés. Se dedicó preferentemente a la teoría de ecuaciones, dominio en el que encontró diversos resultados, entre los que destaca el reconocido teorema que lleva su nombre formulado en 1691. Su obra más importante fue un Tratado de Álgebra en la que inventó la notación para representar la raíz n-ésima de un número x.

 **Enunciado**

Sea $a < b$ y suponga que $f : [a, b] → {\rm I\!R}$ es derivable en $(a, b)$ y continua
en $[a, b]$ y $f(a) = f(b)$. 

Entonces $∃ x_0 ∈ (a, b)$ tal que  $f'(x_0) = 0$

Traduciremos en enunciado anterior en lo siguiente:

Si una función es contínua en un intervalo $[a,b]...$
1.   Es derivable en su interior $(a,b)$, es decir, no presenta picos.
2.   Tiene el mismo valor en los extremos del intervalo, es decir, $f(a)=f(b)$

...entonces existe al menos un punto de tangente horizontal.

Analicemos el teorema de Rolle en una función cuadrática donde localizaremos las condiciones que el teorema propone.

En la función $f(x)= x^2 + 5$ tenemos claro que es una función continúa y mediante el uso de una gráfica localizaremos las condiciones del teorema...



 ### Derivadas de orden superior
 Sea *f(x)* una función derivable, es posible obtener la k-esima derivada de ducha función. La derivada de orden k es la función que se obtiene al derivar respecto a la variable la función *f(x) k* veces consecutivas. La notación utilizada para diferenciar el orden de la derivada es:
  $$ \frac{d^kf(x)}{dx^k} = f^{(k)}(x)$$

  **Ejemplo**

  Calcular la cuarta derivada de $f(x)= 3x^4 + 2x^2 - 20$ 
  
  Para calcular la derivada solicitada de *f(x)* se deben realizar los calculos recursivos desde la primer derivada de la función.
  $$f´(x)= 12x^3 + 4x$$ 
  $$f´´(x)= 36x^2 + 4$$ 
  $$f´´´(x)= 72x $$ 
  $$f^{(4)}(x)= 72 $$ 

 ### Localización de puntos máximos y mínimos relativos, regiones de concavidad y puntos de inflexión 
 Para abordar la problemática de encontrar los puntos extremos es importante la definición de algunos conceptos:


*   **Máximo absoluto:** Es un número $f(c_1)$ de una función tal que $f(x)\leq f(c_1)$ para toda *x* en el dominio de *f*
*   **Mínimo absoluto:** Es un número $f(c_2)$ de una función tal que $f(x)\geq f(c_2)$ para toda *x* en el dominio de *f*
*   **Máximo relativo:** Es un número $f(c_1)$ de una función tal que $f(x)\leq f(c_1)$ para toda *x* en algún intervalo abierto que contiene a $c_1$.
*   **Mínimo relativo:** Es un número $f(c_2)$ de una función tal que $f(x)\geq f(c_1)$ para toda *x* en algún intervalo abierto que contiene a $c_1$.
*   **Punto crítico:** El punto crítico de una función *f* es un número *c* en su dominio para el cual $f´(c)=0$ o $f´(c)$ no existe.
*   **Cóncavidad:** Si *f* es una función diferenciable y $f´$ es una función *creciente* en *(a,b)* se dice que la función es cóncava.
*   **Convexidad:** Si *f* es una función diferenciable y $f´$ es una función *decreciente* en *(a,b)* se dice que la función es convexa.
*   **Punto de inflexión:** Si *f* es continua sobre un intervalo *(a,b)* que contiene al número *c* se llama punto de inflexión a $(c,f(c))$si en dicho punto hay una recta tangente y la gráfica cambia de concavidad en este punto. Suponiendo que $(c,f(c))$  es punto de inflexión para *f*, entonces $f´´(c)=0$ o $f´´(c)$ no existe.

No todas las funciones tiene máximo o mínimo absoluto como ejemplo tenemos a las funciones lineales las cuales no tienen dichos puntos. Algunas veces es sencillo identificar graficamente estos puntos.

#### Teorema de la unicidad del punto crítico
Suponiendo que *c* es el **único** punto crítico de una función *f* dentro de un intervalo. Si se demuestra que *f(c)* es un extremo relatico, entonces *f(c)* es un extremo absoluto.

#### Criterio de la primera derivada
Sea *f* una función continua sobre *[a,b]* y diferenciable sobre *(a,b)*  excepto posiblemente en *c* punto crítico, entonces *f(c)* puede clasificarse como sigue.

1. Si $f'(x)$ cambia de negativa a positiva en c, entonces $f(c)$ es un mínimo relativo.

2. Si $f'(x)$   cambia de positiva a negativa en c, entonces $f(c)$ es un máximo relativo.

 3. Si $f'(x)$   es positiva en ambos lados de c o negativa en ambos lados de c, entonces $f(c)$ no es ni un mínimo ni un máximo relativo. 


#### Teorema del valor extremo
Existe un teorema el cual facilita conocer si una función tiene un máximo y un mínimo absoluto, pues en caso de tener una función continua en un intervalo cerrado aseguramos que existen $f(c_1)$ y $f(c_2)$ tales que $f(c_1)\geq f(x)\leq f(c_2)$ llamados el mínimo y el máximo respectivamente.

**Enunciado**

Una función *f* continua sobre un intervalo cerrado *[a,b]* siempre tiene un máximo absoluto y un mínimo absoluto sobre el inteervalo.

#### Criterio de la segunda derivada

 Sea *f* una función con su primera derivada definida, al menos, en un intervalo abierto *(a,b)* conteniendo al número c. Si $f´´$ esta definida entonces podemos considerar los siguiente aspectos: 

1.  Si $f´´(c) < 0$  entonces se dice que *f(c)* es un máximo relativo.

2.  Si $f´´(c) > 0$ entonces se dice que *f(c)* es un mínimo relativo. 

#### Teorema de funciones crecientes y decrecientes
Es posible conocer si una función es creciente o decreciente en un intervalo mediante el uso de derivadas.

**Enunciado**

Sea *f* una función continua sobre *[a,b]* y diferenciable sobre $(a,b)$.


1.   Si $f´(x) > 0$ para toda *x* en *(a,b)*, entonces la función *f* es *creciente* sobre el intervalo $[a,b]$
2.   Si $f´(x) < 0$ para toda *x* en *(a,b)*, entonces la función *f* es *decreciente* sobre el intervalo $[a,b]$

#### Teorema de concavidad
Para conocer la concavidad de una función es de gran utilidad el uso del siguiente teorema.

**Enunciado**


Sea *f* una función para la cual la segunda derivada existe sobre un intervalo *(a,b)*


1.   Si $f´´(x) > 0$ para toda *x* en *(a,b)* entonces la gráfica de *f* es    cóncava sobre *(a,b)*.
2.    Si $f´´(x) < 0$ para toda *x* en *(a,b)* entonces la gráfica de *f* es    convexa sobre *(a,b)*. 






## Ejecicios sobre derivadas

1.   Utilizando la regla de L´Hopital encuentra el límite de $f(x)=\frac{sen(x)}{x}$ cuando x tiende a cero.

 **Solución**

 Al cumplirse las condiciones de la regla podemos asegurar que:
 $$ \lim_{x \to 0} \frac{sen (x)}{x} = \lim_{x \to 0} \frac{sen´(x)}{x´} = \lim_{x \to 0} \frac{cos(x)}{1}=1$$

2.  Encuentra la primera y segunda derivada de la siguientes funciones:
$$  f(x)= x^5 - 2x^3 + x$$ 
Aplicando las reglas de derivación obtenemos:
  $$f´(x)= 5x^4 - 6x^2 +1$$ 
  $$f´´(x)= 20x^3 - 12 x $$ 

3.  Encuentra la primera y segunda derivada de la siguientes funciones:
$$  f(x)= 4 cos x^2$$ 
Aplicando las regla de la cadena obtenemos:
  $$f´(x)= -4 senx^2(2x) = -8x senx^2$$ 
  $$f´´(x)= -8 senx^2 -8x (cosx^2(2x)) = -8 senx^2 -16x^2 (cosx^2)$$ 


4.   Encuentra el mínimo y máximo relativo de la siguiente función:
$$f(x)= 4x^4 - 4x^2$$

 **Solución**

 Graficaremos la función para poder visualizarla y observar que *f* es una función simétrica.








```python
def f(x):  
  return 4*x**4 - 4*x**2

x = np.linspace(-2, 4, num=100)

par = [{'title':'Grafico de $f(x)= 4x^4 - 4x^2$', 
        'xlabel':'$x$',
        'ylabel':'$y=f(x)$',
       'ylim':(-2,4), 'xlim':(-2,2)}
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x= x, y= f(x))
graf.legend()
graf.show()
```

Utilizando el Criterio de la segunda derivada es posible obtener su promera y segunda derivada.
$$f(x)= 4x^4 - 4x^2$$
$$f´(x)= 16x^3 - 8x$$
$$f´´(x)= 48x^3 - 8$$

Utilizando la definición de punto crítico es posible observar que los valores 0, $ \frac{\sqrt2}{2}$, $ - \frac{\sqrt2}{2}$ pues 
$$f´(0)= 16(0)^3 - 8(0)=0$$
$$f´\left(- \frac{\sqrt2}{2}\right)= 16\left(- \frac{\sqrt2}{2}\right)^3 - 8\left(- \frac{\sqrt2}{2}\right)=0$$
$$f´\left( \frac{\sqrt2}{2}\right)= 16\left( \frac{\sqrt2}{2}\right)^3 - 8\left( \frac{\sqrt2}{2}\right)=0$$

Por lo que al evaluar dichos puntos en la segunda derivada obtenemos

| $x$        | $Signo$ |  f(x) | $Punto$ |
| :-:  | :-: | :-: |  :-: |
| 0  | - |0 | máximo relativo |
| $ \frac{\sqrt2}{2}$ | +	|-1	| mínimo relativo|
| $ - \frac{\sqrt2}{2}$	| + |	-1 |	mínimo relativo|

Es posible observar que el punto de inflexión de $f(x)= 4x^4 - 4x^2$ se encuentra en $\pm \frac{\sqrt6}{6}$  pues es simétrica.

Ubicando los puntos en la gráfica...

## Solución del ejemplo de interés
Para darle solución al problema inicial sobre el estudiante que olvida las palabras estudiadas según el comportamiento de la función $f(t)$.

Es necesario el cálculo de la derivada utilizando las herramientas ya presentadas en la sección de teoría.

Para el cálculo de la drivada
$$ f(t)= 80(1-0.1t)^2$$
Utilizando reglas de derivación:
$$ f´(t)= 80\cdot 2 \cdot (1-0.1t) \cdot (-0.1)$$
$$ =  160 \cdot (1-0.1t) \cdot (-0.1)$$
$$ =  -16 \cdot (1-0.1t)$$

Visualicemos las funciones anteriores en una gráfica:




```python
def x_1(x):  
  return 80*(1-0.1*x)**2

def x_c1(x):  
  return -16 * (1-0.1*x)


x = np.linspace(-1, 25, num=100)

par = [{'title':'Grafico de $f(x)= 4x^4 - 4x^2 y derivada$', 
        'xlabel':'$x$',
        'ylabel':'$y=f(x)$',
       'ylim':(-25,170), 'xlim':(0,25)}
      ]

graf = vis.planoCartesiano(par=par)
graf.plot(x= x, y= x_1(x), par={'label':'Función $f(x)= 80(1-0.1x)^2$'})
graf.plot(x= x, y= x_c1(x), par={'label':'Función $f(x)= -16 * (1-0.1*x)$'})
graf.legend()
graf.show()
```

Para obtener la interpretación deseada por el estudiante basta con evaluar la derivada en el punto deseado.
$$ f´(x) =  -16 \cdot (1-0.1t)$$
$$ f´(2) =  -16 \cdot (1-0.1\cdot2)$$
$$ f´(2) = -12.8$$

Por lo que después de segundo día de haber estudiado, olvidará 12.8 palabras cada día. Lo cual representa la razon de cambio según olvidará las palabras el estudiante.

Por último se verificarán los resultados en Python.



```python
#Derivada
x = Symbol('x')
fx = 80*(1-0.1*x)**2
diff(fx, x)
```

```python
#Resultado
diff(fx, x).subs(x, 2)
```
